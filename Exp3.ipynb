```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import fetch_openml
boston = fetch_openml(name="Boston", version=1, as_frame =True)

X=boston.data
Y=boston.target.astype(float)
X_rm = X[["RM"]]

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X_rm, Y, test_size=0.2, random_state=42)

from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)

from sklearn.metrics import r2_score
r2 = r2_score(y_test, y_pred)
print(f"R2 SCORE: {r2:.4f}")

plt.figure(figsize=(6,6))
plt.scatter(x_test, y_test, label='Actual Prices')
plt.plot(x_test, y_pred, color='black', linewidth=2, label='Predicted')
plt.xlabel('RM')
plt.ylabel('Price')
plt.title('Boston Housing Prices Prediction')
plt.grid(True)
plt.legend()
```

```python
#This is for all the 14 Features
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

#Regularixation
from sklearn.preprocessing import StandardScaler
std = StandardScaler()
X_train_scaled = std.fit_transform(X_train)
X_test_scaled = std.transform(X_test)

#Ridge Regression
from sklearn.linear_model import Ridge

ridge = Ridge(alpha=1.0)
ridge.fit(X_train_scaled, Y_train)

y_pred_ridge = ridge.predict(X_test_scaled)

r2_ridge = r2_score(Y_test, y_pred_ridge)
print("R2 score (Ridge):", r2_ridge)


# Lasso Regression (L1-Regularization)
from sklearn.linear_model import Lasso

lasso = Lasso(alpha=0.1)

lasso.fit(X_train_scaled, Y_train)

y_pred_lasso = lasso.predict(X_test_scaled)

r2_lasso = r2_score(Y_test, y_pred_lasso)
print("R2 score (Lasso):", r2_lasso)
```

```python
median_price = Y.median()

y_test_class = (y_test >= median_price).astype(int)
y_pred_class = (y_pred >= median_price).astype(int)

from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

cm = confusion_matrix(y_test_class, y_pred_class)
print(cm)

disp = ConfusionMatrixDisplay(
    confusion_matrix=cm,
    display_labels=["Low Price", "High Price"]
)

disp.plot()
plt.title("Confusion Matrix for Housing Price Classification")
plt.show()
```
